```python
import numpy as np
```

# 1.Initialize Vocab

```python
# some text with lots of repetitions
text = 'like liker love lovely hug hugs hugging hearts'

chars = list(set(text))
chars.sort() # initial vocab is sorted

for l in chars:
  print(f'"{l}" appears {text.count(l)} times.')
```

```python
# make a vocabulary
vocab = { word:i for i,word in enumerate(chars) }
vocab
```

```python
# the text needs to be a list, not a string
# each element in the list is a token
origin_tokens = list(text)
print(text)
print(origin_tokens)
```

# 2. Byte Pair Encoding

```python
from typing import List, Dict
def get_pair_stats(tokens: List) -> Dict:
  token_pairs = {}

  for i in range(len(tokens)-1):
    # create pair
    pair = tokens[i] + tokens[i+1]
    # frequencies
    if pair in token_pairs:
      token_pairs[pair] += 1
    else:
      token_pairs[pair] = 1
  return token_pairs

def update_vocab(token_pairs: Dict, vocab: Dict):
  idx = np.argmax(list(token_pairs.values()))
  newtoken = list(token_pairs.keys())[idx]

  vocab[newtoken] = len(vocab)
  return vocab, newtoken

def generate_new_token_seq(prevTokens: List, newtoken) -> List:
  newTokens = []
  i = 0
  while i < len(prevTokens) - 1:
    if prevTokens[i] + prevTokens[i+1] == newtoken:
      newTokens.append(newtoken)
      i += 2
    else:
      newTokens.append(prevTokens[i])
      i += 1

    # add the final token to the list (missing in the video)
  if i < len(prevTokens):
    newTokens.append(prevTokens[i])

  return newTokens
```

```python
vocab_size = 25
updated_tokens = origin_tokens.copy()

while len(vocab) < vocab_size:
  pair = get_pair_stats(updated_tokens)
  vocab, newtoken = update_vocab(pair, vocab)
  updated_tokens = generate_new_token_seq(updated_tokens, newtoken)
  print(f'Vocab has {len(vocab)} tokens after adding "{newtoken}"')
```
